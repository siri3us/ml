```python
from ml.generators import BatchGenerator

arrays = [np.arange(10), np.arange(10)[::-1]]
gen = BatchGenerator(*arrays, batch_size=5, shuffle=True, infinite=True, 
                     fixed_batch_size=False, random_state=0)
n = 0
for a, b in gen:
    print(a.shape, a, b.shape, b)
    n += 1
    if n >= 100:
        break
```
