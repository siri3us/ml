# Установка Anaconda в Linux

### Базовая установка

Скачать подходящий дистрибутив anaconda с официального сайта https://www.anaconda.com/, или напрямую по ссылке https://repo.continuum.io/archive/Anaconda3-5.0.1-Linux-x86_64.sh (python 3.6, x86-64)

Запустить процесс установки выполнив в консоли команду
> `$ bash Anaconda-latest-Linux-x86_64.sh`

По умолчанию установщик должен предолжить вам установку в директорию `$HOME/anaconda3` (`$HOME` &mdash; ваша домашняя директория), а также прописать изменения в соответствующие переменные окружения `PATH` (иногда в `PYTHONPATH`), так что после установки в скрипте `$HOME/.bashrc` должна появиться следующая строка:
> `$ export PATH="$HOME/anaconda3/bin:$PATH"`

Наличие этой строки можно проверить командой 
> `cat $HOME/.bashrc$`

или

> `cat $HOME/.bashrc$ | grep PATH`

Чтобы изменения, внесенные в `.bashrc` вступили в силу, перезапустите консоль. После чего содержание переменной окружения `$PATH`, выводимое командой
> `echo $PATH`

дожно иметь приблизительно следующий вид
> `$HOME/anaconda3/bin:/usr/local/sbin:/usr/local/bin:/usr/sbin:...`


Если вдруг в результате установки в переменной `$PATH` не содержится путь к анаконде, т.е. `$HOME/anaconda3/bin`, что возможно, если права доступа пользователя, из под которого происходит установка, не позволяют внести изменения в файл `$HOME/.bashrc`, то внесите соответствующие изменения вручную (возможно из под суперпользователя). Вместо редактирования файла `.bashrc` можно добавить строку 
> `$ export PATH="$HOME/anaconda3/bin:$PATH"`

в файл `.profile`, однако в этом случае для вступления изменений в силу потребуется повтороно залогиниться в систему.


### Создание отдельного окружения 
На данном этапе у вас уже должна стоять установленная версия анаконды. Теперь создадим именованное окружение с именем `amsl`, основной версией питона в котором будет 3.5
> `$ conda create -n amsl python=3.5`

По умолчаню после установки ваше окружение называется `root`. После выполнения команды выше у вас появиться второе окружение `amsl`. Теперь в результате выполнения команды
> `$ conda info --envs`

должен выводиться список из двух окружений. Переключение между окружениями выполняется командами
> `$ source activate amsl`

При этом командная строка должна принять вид
> `(amsl)$`

Чтобы переключиться обратно на `root`, соответсвенно, требуется выполнить
> `(amsl)$ source activate root`

Окружения позволяют устанавливать в системе конфликтующие между собой библиотеки посредством помещения их в отдельные окружения.

### Библиотеки
Перейдем в окружение `amsl`. 
> `$ source activate amsl`

Выполним установку необходимых библиотек

> `$ pip install numpy scipy` # библиотеки для работы с матрицами и случайными величинами <br>
> `$ pip install matplotlib`  # библиотека для работы с графикой</br> 

> `$ pip install scikit-learn` # библиотека алгоритмов машинного обучения<br>
> `$ pip install scikit-image` # библиотека для работы с изображениями (опционально)</br>


### Jupyter notebook
По идее после установки анаконды и перехода в соответствующее вашей задаче окружение, выполнение команды
> `$(amsl) jupyter notebook`

дожно запустить страницу в браузере для работы с ноутбуками `.ipynb`. Если этого вдруг не произошло, то запустите установку `jupyter notebook`:
> `$(amsl) pip install jupyter notebook`
 
Иногда jupyter notebook не позволяет при создании нового файла `.ipynb` выбрать для него kernel (окружение для запуска) из списка доступных в системе. В частности, окружение `amsl` может быть недоступно про попытке создать новый файл, а доступны только `root` и `default`. В этом случае выполните команды
> `$pip install jupyter notebook`<br>
> `$conda install ipykernel`</br>

из под каждого доступного окружения.
